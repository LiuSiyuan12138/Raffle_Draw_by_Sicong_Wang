```python
import requests
from pyquery import PyQuery as pq
import pandas as pd
headers = {
    'Host': 'm.weibo.cn',
    'Referer': 'https://m.weibo.cn/u/2480489811?uid=2480489811&luicode=10000011&lfid=1076031892723783&featurecode=20000320',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/67.0.3396.87 Safari/537.36',
    'X-Requested-With': 'XMLHttpRequest',
    'X-DevTools-Emulate-Network-Conditions-Client-Id': 'A20EA5B172E6DC82709D213A40AD0E8F'
}


def get_page(page):
    url = 'https://m.weibo.cn/api/container/getIndex?uid=2480489811&luicode=10000011&lfid=1076031892723783&featurecode=20000320&type=uid&value=2480489811&containerid=1076031892723783&page=%d' % page
    try:
        res = requests.get(url, headers=headers)
        if res.status_code == 200:
            return res.json()
    except requests.ConnectionError as e:
        print("Error", e.args)


def parse_page(json):
    if json:
        items = json.get('data').get('cards')
        for item in items:
            item = item.get('mblog')
            weibo = {}
            weibo['text'] = pq(item.get('text')).text()
            weibo['attitudes'] = item.get('attitudes_count')
            weibo['comments'] = item.get('comments_count')
            weibo['reposts'] = item.get('reposts_count')
            weibo['time']=item.get('created_at')
            yield weibo


if __name__ == '__main__':
    text=[]
    attitudes=[]
    comments=[]
    reposts=[]
    times=[]
    dic={}
    for page in range(1, 12):
        json = get_page(page)
        results = parse_page(json)
        for res in results:
            text.append(res['text'])
            attitudes.append(res['attitudes'])
            comments.append(res['comments'])
            reposts.append(res['reposts'])
            times.append(res['time'])
    dic={'text':text,'attitudes':attitudes,'comments':comments,'reposts':reposts,'times':times}
    data=pd.DataFrame(dic)
```
