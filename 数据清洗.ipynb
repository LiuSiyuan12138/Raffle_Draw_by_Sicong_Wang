```python
import pandas as pd
import numpy as np
```

```python
df_list=[]
for i in range(1,10):
    file_name='loser_sample//loser'+str(i)+'.csv'
    df_list.append(pd.read_csv(file_name))
res = pd.concat(df_list,axis=0,ignore_index=True)
```

```python
res.describe()
```

```python
res.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
res
```

```python
def judge_phone(x):
    if 'iPhone' in str(x):
        return 'iPhone'
    elif str(x)=='nan':
        return 'else'
    else:
        return 'Android'
res['source']=res['source'].apply(judge_phone)
```

```python
res
```

```python
res.to_excel('未中奖者抽样.xlsx')
```
